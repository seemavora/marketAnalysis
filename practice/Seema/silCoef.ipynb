```python
import pandas as pd 
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
%matplotlib inline
```

```python
X = np.random.rand(50,2)
Y = 2+ np.random.rand(50,2)
Z = np.concatenate((X,Y)) #joins a sequence of arrays along a axis
Z = pd.DataFrame(Z)
Z
```

```python
sns.scatterplot(Z[0],Z[1])
```

```python
KMean = KMeans(n_clusters = 2)
KMean.fit(Z)
label = KMean.predict(Z)
# creates two clusters
```

```python
print(f'Silhouette Score(n = 2):{silhouette_score(Z, label)}')
#since the score is closer to 1 it is safe to assume that the clusters are very apart
```

```python
sns.scatterplot(Z[0],Z[1], hue = label)
```

```python
#what about with three clusters
KMean = KMeans(n_clusters = 3)
KMean.fit(Z)
label = KMean.predict(Z)
print(f'Silhouette Score(n= 3): {silhouette_score(Z, label)}')
sns.scatterplot(Z[0], Z[1], hue = label, palette = 'inferno_r')
```
